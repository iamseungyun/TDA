```python
from DTM_filtrations import *

import gudhi
import numpy as np
import matplotlib.pyplot as plt
```

## First Dataset

```python
N_obs = 150
N_out = 50
X = SampleOnCircle(N_obs, N_out, is_plot=True)
```

```python
#Usual Rips complex
st_rips = gudhi.RipsComplex(points=X).create_simplex_tree(max_dimension=2)
diagram_rips = st_rips.persistence()

gudhi.plot_persistence_diagram(diagram_rips)
plt.title('Persistence diagram of the Rips complex')
```

```python
#Compute DTM
m = 0.1
DTM_values = DTM(X,X,m)

plot = plt.scatter(X[:,0], X[:,1], c=DTM_values)
plt.colorbar(plot)
plt.axis('equal')
plt.title('Values of the DTM on X with parameter m=' +str(m))
```

```python
#DTM-based Rips complex
p = 1
dimension_max = 2
st_DTM = DTMFiltration(X,m,p,dimension_max)
diagram_DTM = st_DTM.persistence()

gudhi.plot_persistence_diagram(diagram_DTM)
plt.title('Persistence diagram of the DTM-based filtration with parameter p='+str(p))
```

```python
from gudhi.dtm_rips_complex import *

m = 0.1
N = np.shape(X)[0]
k = int(m*N)

dtm_rips = gudhi.dtm_rips_complex.DTMRipsComplex(points=X, k=k)
st_DTM = dtm_rips.create_simplex_tree(max_dimension=2)
diagram_DTM = st_DTM.persistence()

gudhi.plot_persistence_diagram(diagram_DTM)
plt.title('Persistence diagram of the DTM-filtration with parameter p='+str(p))
```

```python
P = [2, 6.62, np.inf]

for p in P:
    st =DTMFiltration(X,m,p,dimension_max=2)
    diagram = st.persistence()
    
    plt.figure()
    gudhi.plot_persistence_diagram(diagram)
    plt.title('Persistence diagram of the DTM-filtration with parameter p='+str(p))
```

## Second Dataset

```python
N_obs = 350
N_out = 0
X = SampleOnNecklace(N_obs, N_out, is_plot=True)
```

```python
#Usual alpha complex
st_alpha = gudhi.AlphaComplex(points=X).create_simplex_tree()
diagram_alpha = st_alpha.persistence()

gudhi.plot_persistence_diagram(diagram_alpha)
plt.title('Persistence diagram of the alpha-complex')
```

```python
#DTM
m = 0.05
DTM_values = DTM(X,X,m)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

plot = ax.scatter(X[:,0], X[:,1], X[:,2], c=DTM_values)
fig.colorbar(plot)
ax.view_init(30,30)

m = 0.3
DTM_values = DTM(X,X,m)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

plot = ax.scatter(X[:,0], X[:,1], X[:,2], c=DTM_values)
fig.colorbar(plot)
ax.view_init(30,30)
```

```python
#DTM-based alpha complex
m = 0.05
p = 1
dimension_max = 3

simplex_tree_DTM_alpha = AlphaDTMFiltration(X,m,p,dimension_max)
diagram_DTM_alpha = simplex_tree_DTM_alpha.persistence()

gudhi.plot_persistence_diagram(diagram_DTM_alpha)
plt.title('Persistence diagram of the Alpha-DTM-filtration with parameter p='+str(p))
```

```python
#DTM-based alpha complex
m = 0.3
p = 1
dimension_max = 3

simplex_tree_DTM_alpha = AlphaDTMFiltration(X,m,p,dimension_max)
diagram_DTM_alpha = simplex_tree_DTM_alpha.persistence()

gudhi.plot_persistence_diagram(diagram_DTM_alpha)
plt.title('Persistence diagram of the Alpha-DTM-filtration with parameter p='+str(p))
```

```python
m = 0.03
P = [2, 6.62, np.inf]
dimension_max = 3

for p in P:
    st = AlphaDTMFiltration(X,m,p,dimension_max)
    diagram = st.persistence()
    
    plt.figure()
    gudhi.plot_persistence_diagram(diagram)
    plt.title('Persistence diagram of the DTM-filtration with parameter p='+str(p))
```
